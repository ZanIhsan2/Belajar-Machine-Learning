```python
import pandas as pd

# Baca dataset
df = pd.read_csv("../data/day.csv")

# Acak urutan data (supaya tidak bias)
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Hitung ukuran dataset
n_total = len(df)
n_test = int(0.2 * n_total)      # 20% untuk testing
n_train = n_total - n_test       # sisanya untuk training
n_val = int(0.1 * n_train)       # 10% dari training untuk validation
n_train = n_train - n_val        # sisanya training asli

# Bagi dataset
train = df.iloc[:n_train]
val = df.iloc[n_train:n_train+n_val]
test = df.iloc[n_train+n_val:]

# Tampilkan jumlah data
print("Jumlah total data:", n_total)
print("Jumlah data Training:", len(train))
print("Jumlah data Validation:", len(val))
print("Jumlah data Testing:", len(test))

# Tampilkan 5 baris teratas dari masing-masing
print("\nData Training:")
print(train.head())

print("\nData Validation:")
print(val.head())

print("\nData Testing:")
print(test.head())
```
